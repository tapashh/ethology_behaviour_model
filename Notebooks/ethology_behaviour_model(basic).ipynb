```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# The 20 dimensions defined by the Evolutionary Ethology Framework
DIMENSIONS = [
    "Dominance", "Agreeableness", "Curiosity", "Strategic Thinking",
    "Emotional Stability", "Creativity", "Social Dependence", "Independence",
    "Long-term Planning", "Impulsiveness", "Empathy", "Competitiveness",
    "Discipline", "Adaptability", "Learning Speed", "Risk Taking",
    "Communication Style", "Cooperation", "Ambition", "Persistence"
]
```

```python
# Fixed Centroids for Animal Archetypes
# Values represent the normalized 0-100 scale across the 20 dimensions
ARCHETYPES = {
    # Primates
    "Chimpanzee": np.array([95, 20, 50, 85, 40, 60, 70, 30, 60, 75, 20, 95, 60, 70, 60, 80, 70, 30, 90, 75]),
    "Bonobo":     np.array([15, 95, 60, 50, 70, 65, 85, 20, 40, 50, 95, 10, 40, 80, 60, 30, 85, 95, 30, 50]),
    
    # Apex Predators
    "Wolf":       np.array([70, 60, 40, 80, 85, 40, 90, 20, 85, 20, 60, 65, 95, 40, 50, 50, 60, 95, 60, 95]),
    "Cheetah":    np.array([60, 40, 65, 50, 40, 70, 20, 90, 20, 95, 30, 70, 30, 60, 60, 95, 30, 20, 75, 20]),
    
    # Producers & Innovators
    "Raven":      np.array([30, 50, 95, 75, 60, 95, 30, 85, 60, 60, 50, 40, 40, 85, 95, 60, 50, 40, 50, 50]),
    "Octopus":    np.array([40, 10, 90, 95, 70, 100, 10, 100, 60, 30, 10, 30, 70, 95, 90, 40, 10, 10, 40, 60]),
    
    # Conservers
    "Vampire Bat":np.array([20, 85, 40, 50, 60, 40, 95, 10, 50, 40, 95, 10, 60, 50, 50, 30, 70, 95, 20, 60])
}
```

```python
def classify_archetype(user_vector, centroids=ARCHETYPES, beta=6000):
    """
    Computes possibilistic membership of a user vector to fixed archetype centroids.
    
    Parameters:
    user_vector (np.array): 20-dimensional array of the user's scores.
    centroids (dict): Dictionary containing archetype names and their centroid arrays.
    beta (int): Tuning parameter for distance decay. Lower = stricter matching.
    
    Returns:
    dict: Archetypes and their percentage match, sorted descending.
    """
    memberships = {}
    
    for name, centroid in centroids.items():
        # Calculate squared Euclidean distance: ||x_j - v_i||^2
        distance_sq = np.sum((user_vector - centroid) ** 2)
        
        # Possibilistic decay formula
        match_percentage = np.exp(-distance_sq / beta) * 100
        memberships[name] = round(match_percentage, 2)
        
    # Sort by highest percentage match
    sorted_memberships = dict(sorted(memberships.items(), key=lambda item: item[1], reverse=True))
    
    return sorted_memberships
```

```python
# Profile 1: The Startup Founder
# High Risk (92), Impulsive (85), Creativity (90), Low Persistence (30), Low Discipline (40)
startup_founder = np.array([
    60,  # Dominance 
    50,  # Agreeableness
    70,  # Curiosity
    60,  # Strategic Thinking
    45,  # Emotional Stability
    90,  # Creativity
    40,  # Social Dependence
    80,  # Independence
    40,  # Long-term Planning
    85,  # Impulsiveness
    45,  # Empathy
    65,  # Competitiveness
    40,  # Discipline
    80,  # Adaptability
    75,  # Learning Speed
    92,  # Risk Taking
    60,  # Communication Style
    40,  # Cooperation
    70,  # Ambition
    30   # Persistence
])

# Execute model
results = classify_archetype(startup_founder)

# Display results
print("Possibilistic Archetype Matches:")
print("-" * 30)
for archetype, score in results.items():
    if score > 20: # Display threshold
        print(f"{archetype:<15} {score}% Match")
```
